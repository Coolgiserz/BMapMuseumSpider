### What's the meaning of status?

```python
import requests
import urllib3
def test_url(url,header):
    req = requests.get(url,headers=header,allow_redirects=False)
    print(req)
    code = req.status_code
    print('''
        URL:{0}
        请求响应码：{1}, 
        
    '''.format(url,code))
    if code==200:
        return True
    else:
        return False
```
